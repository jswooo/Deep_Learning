### 3.2.4 시그모이드 함수

```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))
```

#### 확인

```python
x = np.array([-1., 1., 2.])
sigmoid(x)
```

#### 시각화 

```python
x = np.arange(-5., 5., 0.1)
y = sigmoid(x)
plt.plot(x, y)
plt.ylim(-0.1, 1.1)
plt.show()
```

활성화함수는 항상 비선형 함수여야 함.  
선형 함수를 사용하면 신경망에서 여러 층으로 쌓는게 의미없게 됌.  
h(x) = cx, h(h(h(x)))) = c * c * c * x = ax => 1층과 같음

### 3.2.7 ReLU 구현

```python
def relu(x):
    return np.maximum(0, x)
```

#### 시각화

```python
x = np.arange(-1., 1., 0.1)
y = relu(x)
plt.plot(x, y)
plt.ylim(-0.1, 1.1)
plt.show()
```

## 3.4 3층 신경망 구현 (p.85)

```python
# 입력층에서 1층으로 신호 전달
X = np.array([1., 0.5]) # input, 1x2 
W1 = np.array([[0.1, 0.3, 0.5], [0.2, 0.4, 0.6]]) # 가중치, 2x3
B1 = np.array([0.1, 0.2, 0.3]) # bias, 1x3 

A1 = np.dot(X, W1) + B1
Z1 = sigmoid(A1) # 활성화함수를 거친 최종 결과물

print(A1)
print(Z1)
```

```python
# 1층에서 2층으로 신호 전달
W2 = np.array([[0.1, 0.4], [0.2, 0.5], [0.3, 0.6]]) # 가중치, 3x2
B2 = np.array([0.1, 0.2]) # bias, 1x2

A2 = np.dot(Z1,W2) + B2 # 1층의 출력을 input으로 활용
Z2 = sigmoid(A2)

print(A2)
print(Z2)
```

```python
# 2층에서 출력층으로 신호 전달

# 최종 출력을 위한 활성화 함수인 '함등함수' 정의
def identity_function(x):
    return x

W3 = np.array([[0.1, 0.3], [0.2, 0.4]])
B3 = np.array([0.1, 0.2])

A3 = np.dot(Z2,W3) + B3
Y = identity_function(A3) # 최종 출력값

print(A3)
print(Y)
```

#### 구현 정리 

```python
def init_network():
    network = dict()
    network['W1'] = np.array([[0.1, 0.3, 0.5], [0.2, 0.4, 0.6]])
    network['b1'] = np.array([0.1, 0.2, 0.3])
    network['W2'] = np.array([[0.1, 0.4], [0.2, 0.5], [0.3, 0.6]])
    network['b2'] = np.array([0.1, 0.2])
    network['W3'] = np.array([[0.1, 0.3], [0.2, 0.4]])
    network['b3'] = np.array([0.1, 0.2])
    
    return network

# 순전파
def forward(network, x):
    W1, W2, W3 = network['W1'], network['W2'], network['W3'] # weights
    b1, b2, b3 = network['b1'], network['b2'], network['b3'] # bias 
    
    # 입력 - 1층
    a1 = np.dot(x,W1) + b1
    z1 = sigmoid(a1)
    
    # 1층 - 2층 
    a2 = np.dot(z1,W2) + b2
    z2 = sigmoid(a2)
    
    # 2층 - 3층 
    a3 = np.dot(z2,W3) + b3
    y = identity_function(a3)
    
    return y

network = init_network()
x = np.array([1., 0.5]) # 초기값
y = forward(network, x)
print(y)
```

## 3.5 출력층 설계

일반적으로 회귀에서는 `항등함수`를, 분류에서는 `softmax 함수`를 사용 

```python
# softmax 구현
a = np.array([0.3, 2.9, 4.0])

exp_a = np.exp(a) # 지수 함수(분자)
print(exp_a)

sum_exp_a = np.sum(exp_a) # 지수 함수의 합(분모)
print(sum_exp_a)

y = exp_a / sum_exp_a # softmax 확률 값
print(y)
```

#### softmax overflow 문제

지수 형태라서 너무 큰 값을 갖게 됌

```python
a = np.array([1010, 1000, 990])
print(np.exp(a) / np.sum(np.exp(a))) # overflow로 nan 출력

c = np.max(a) # 최대값을 빼서 overflow 방지
output = np.exp(a-c) / np.sum(np.exp(a-c)) 
print(output)
print(np.sum(output))
```

#### 최종 softmax 함수

```python
def softmax(a):
    c = np.max(a)
    exp_a = np.exp(a-c)
    sum_exp_a = np.sum(exp_a)
    y = exp_a / sum_exp_a
    
    return y
```

```python
# softmax의 합은 1
a = np.array([0.3, 2.9, 4.])
y = softmax(a)
print(y)
print(np.sum(y))
```

#### 신경망을 통한 추론

```python
import pickle
import time

def get_data():
    (x_train, t_train), (x_test, t_test) = load_mnist(flatten=True, normalize=True, one_hot_label=False)
    return x_test, t_test


# 학습된 가중치 활용하여 추론 진행 
def init_network():
    with open("sample_weight.pkl", "rb") as f:
        network = pickle.load(f)
    
    return network


# 추론
def predict(network, x):
    W1, W2, W3 = network['W1'], network['W2'], network['W3']
    b1, b2, b3 = network['b1'], network['b2'], network['b3']
    
    a1 = np.dot(x, W1) + b1
    z1 = sigmoid(a1)
    a2 = np.dot(z1, W2) + b2
    z2 = sigmoid(a2)
    a3 = np.dot(z2, W3) + b3
    y = softmax(a3)
    
    return y


x, t = get_data()
network = init_network()

start = time.time()

accuracy_cnt = 0 # 정확도 계산용
for i in range(len(x)):
    y = predict(network, x[i])
    p = np.argmax(y) # 확률이 가장 높은 원소의 인덱스 리턴
    if p == t[i]:
        accuracy_cnt += 1

end = time.time()
        
print("Accuracy : " + str(float(accuracy_cnt)/len(x)))
print("time : " + str(end-start))
```

#### 배치 처리

수치 계산 라이브러리들이 큰 배열 처리에 최적화되어 있으므로, 배치 처리를 하는 것이 이득  
I/O를 줄이고, CPU나 GPU 연산 비율을 높여서 빠른 계산이 가능

```python
x, t = get_data()
network = init_network()

batch_size = 100 # 배치 크기
accuracy_cnt = 0

start = time.time()

for i in range(0, len(x), batch_size):
    x_batch = x[i: i+batch_size]
    y_batch = predict(network, x_batch)
    p = np.argmax(y_batch, axis=1) # 100x10 에서 1차원(10)의 가장 큰 확률의 인덱스들 가져옴
    accuracy_cnt += np.sum(p==t[i:i+batch_size]) # true, false로 반환 후 sum 

end = time.time()

print("Accuracy : " + str(float(accuracy_cnt)/len(x)))
print("time : " + str(end-start))
```
